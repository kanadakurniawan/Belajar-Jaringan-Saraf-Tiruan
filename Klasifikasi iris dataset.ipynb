```python
#Import required libraries
import keras #library for neural network
import pandas as pd #loading data in table form
import seaborn as sns #visualisation
import matplotlib.pyplot as plt #visualisation
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.preprocessing import normalize #machine learning algorithm library
```

```python
#Reading data
data=pd.read_csv("https://raw.githubusercontent.com/kanadakurniawan/Belajar-Jaringan-Saraf-Tiruan/1d95438528ffade3bde5554b0402cf2c5c262ec9/dataset/Iris.csv")
print("Describing the data: ",data.describe())
print("Info of the data:",data.info())
```

```python
print("10 first samples of the dataset:",data.head(10))
print("10 last samples of the dataset:",data.tail(10))
```

Pengkodean di bawah ini mengubah spesies ke dalam masing-masing kategori untuk dimasukkan ke dalam jaringan saraf

```python
print(data["Species"].unique())
```

```python
data.loc[data["Species"]=="Iris-setosa","Species"]=0
data.loc[data["Species"]=="Iris-versicolor","Species"]=1
data.loc[data["Species"]=="Iris-virginica","Species"]=2
print(data.head())
```

```python
data=data.iloc[np.random.permutation(len(data))]
print(data.head())
```

Mengonversi data ke array numpy agar dapat diproses

```python
X=data.iloc[:,1:5].values
y=data.iloc[:,5].values

print("Shape of X",X.shape)
print("Shape of y",y.shape)
print("Examples of X\n",X[:3])
print("Examples of y\n",y[:3])
```

**Normalisasi**

Dapat dilihat dari atas bahwa fitur dataset pertama memiliki Panjang Kelopak 5.4 cm, Lebar Kelopak 3,9 cm, Panjang Kelopak 1.3 cm, dan Lebar Kelopak 0.4 cm. Namun, rentang dataset mungkin berbeda. Oleh karena itu, untuk mempertahankan akurasi yang baik, fitur setiap dataset harus dinormalisasi ke rentang 0-1 untuk diproses.

```python
X_normalized=normalize(X,axis=0)
print("Examples of X_normalised\n",X_normalized[:3])
```

```python
#Membuat data pelatihan, pengujian, dan validasi
'''
80% -- train data
20% -- test data
'''
total_length=len(data)
train_length=int(0.8*total_length)
test_length=int(0.2*total_length)

X_train=X_normalized[:train_length]
X_test=X_normalized[train_length:]
y_train=y[:train_length]
y_test=y[train_length:]

print("Length of train set x:",X_train.shape[0],"y:",y_train.shape[0])
print("Length of test set x:",X_test.shape[0],"y:",y_test.shape[0])
```

```python
# Neural network module
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
# from keras.layers.normalization import BatchNormalization
from tensorflow.keras.utils import to_categorical

# Change the label to one hot vector
'''
[0] ---> [1 0 0]
[1] ---> [0 1 0]
[2] ---> [0 0 1]
'''
y_train = to_categorical(y_train, num_classes=3)
y_test = to_categorical(y_test, num_classes=3)

print("Shape of y_train", y_train.shape)
print("Shape of y_test", y_test.shape)
```

```python
model=Sequential()
model.add(Dense(1000,input_dim=4,activation='relu'))
model.add(Dense(500,activation='relu'))
model.add(Dense(300,activation='relu'))
# model.add(Dropout(0.2))
model.add(Dense(3,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),batch_size=20,epochs=10,verbose=1)
```

```python
prediction=model.predict(X_test)
length=len(prediction)
y_label=np.argmax(y_test,axis=1)
predict_label=np.argmax(prediction,axis=1)

accuracy=np.sum(y_label==predict_label)/length * 100
print("Accuracy of the dataset",accuracy )
```

Akurasi 96.67% dicapai dalam kumpulan data ini. Dapat dipastikan bahwa untuk setiap periode, jaringan saraf mencoba belajar dari fitur yang ada dan memprediksinya berdasarkan bobot dan biasnya. Untuk setiap periode, bobot dan bias diubah dengan mengurangi rasionya untuk mendapatkan akurasi yang lebih baik setiap saat.
