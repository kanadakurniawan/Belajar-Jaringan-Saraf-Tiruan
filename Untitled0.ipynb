### Apa itu Jaringan Saraf Tiruan (JST)?

Jaringan Saraf Tiruan (JST) adalah sistem komputasi yang terinspirasi oleh cara kerja otak manusia. Tujuan utamanya adalah untuk memproses informasi, belajar dari data, dan membuat keputusan atau prediksi berdasarkan pola yang telah dipelajari. Berikut adalah penjelasan sederhana:

#### 1. Konsep Dasar
- JST terdiri dari **"neuron" buatan**, yang meniru cara kerja neuron di otak manusia.
- Neuron-neuron ini diatur dalam lapisan (layers): **lapisan input**, **lapisan tersembunyi**, dan **lapisan output**.
- Data masuk melalui lapisan input, diproses di lapisan tersembunyi, dan hasil akhirnya keluar melalui lapisan output.

#### 2. Perhitungan Jaringan Saraf Tiruan Sederhana

image.png

Input Layer:
- Input pertama: $x_1 = 2$
- Input kedua: $x_2 = 3$

Hidden Layer:
Bobot ke Neuron 1:
- Dari $x_1$: $w_{11} = 0.6$
- Dari $x_2$: $w_{21} = 1.2$

$$h_1 = (x_1 \cdot w_{11}) + (x_2 \cdot w_{21})$$
$$h_1 = (2 \cdot 0.6) + (3 \cdot 1.2) = 1.2 + 3.6 = 4.8$$

Bobot ke Neuron 2:
- Dari $x_1$: $w_{12} = 0.8$
- Dari $x_2$: $w_{22} = 2.0$

$$h_2 = (x_1 \cdot w_{12}) + (x_2 \cdot w_{22})$$
$$h_2 = (2 \cdot 0.8) + (3 \cdot 2.0) = 1.6 + 6.0 = 7.6$$

Output Hidden Layer:
$$h_1 = 4.8, \ h_2 = 7.6$$

# Output Layer:
Bobot ke Output:
- Dari $h_1$: $w_{h1} = 0.4$
- Dari $h_2$: $w_{h2} = 1.8$

$$y = (h_1 \cdot w_{h1}) + (h_2 \cdot w_{h2})$$
$$y = (4.8 \cdot 0.4) + (7.6 \cdot 1.8) = 1.92 + 13.68 = 15.6$$

# Hasil Akhir:
Output dari hidden layer:
$$h_1 = 4.8, \ h_2 = 7.6$$
Output akhir jaringan:
$$y = 15.6$$


````python
# Perhitungan Jaringan Saraf Tiruan Sederhana
# Input Layer
x1 = 2  # Input pertama
x2 = 3  # Input kedua

# Bobot ke Hidden Layer
w11 = 0.6  # Bobot dari x1 ke neuron 1 di hidden layer
w21 = 1.2  # Bobot dari x2 ke neuron 1 di hidden layer
w12 = 0.8  # Bobot dari x1 ke neuron 2 di hidden layer
w22 = 2.0  # Bobot dari x2 ke neuron 2 di hidden layer

# Aktivasi Neuron di Hidden Layer
h1 = (x1 * w11) + (x2 * w21)  # Output neuron 1 di hidden layer
h2 = (x1 * w12) + (x2 * w22)  # Output neuron 2 di hidden layer

# Bobot ke Output Layer
wh1 = 0.4  # Bobot dari neuron 1 (h1) ke output layer
wh2 = 1.8  # Bobot dari neuron 2 (h2) ke output layer

# Aktivasi Neuron di Output Layer
y = (h1 * wh1) + (h2 * wh2)  # Output akhir jaringan

# Menampilkan hasil
print("Output dari hidden layer (h1, h2):", h1, h2)
print("Output akhir jaringan (y):", y)
````



#### 2. Bagaimana JST Bekerja
- Data diubah menjadi angka (disebut **input**) dan dimasukkan ke dalam jaringan.
- Setiap neuron mengolah input dengan menggunakan rumus matematika sederhana, seperti menjumlahkan nilai dan menerapkan fungsi aktivasi (misalnya, ReLU atau sigmoid).
- Hasil olahan dari satu neuron dikirim ke neuron lain hingga sampai pada output.

#### 3. Proses Belajar
- JST belajar dari data menggunakan **proses pelatihan**, yang melibatkan **forward pass** (memprediksi output) dan **backward pass** (memperbaiki kesalahan dengan mengatur bobot neuron).
- Pelatihan ini dibantu oleh algoritma seperti **backpropagation** dan **optimizers**.

#### 4. Contoh Kehidupan Sehari-Hari
- Saat menggunakan fitur pengenalan wajah di ponsel, JST membantu mengenali wajah Anda.
- Aplikasi seperti Google Translate menggunakan JST untuk memahami dan menerjemahkan bahasa.

#### 5. Analogi Sederhana
- Bayangkan sebuah tim siswa yang bekerja bersama untuk memecahkan masalah.
- Setiap siswa (neuron) memberikan kontribusi berdasarkan keahliannya, saling berkomunikasi, hingga akhirnya menghasilkan jawaban terbaik (output).

#### Kesimpulan
Dengan cara ini, JST bisa digunakan untuk berbagai masalah, seperti memprediksi cuaca, mengenali gambar, atau membuat rekomendasi. Meskipun konsepnya sederhana, implementasinya melibatkan banyak matematika dan teknik komputasi.


#### 6. Rumus Perhitungan Sederhana
Misalkan ada 3 input \(x_1, x_2, x_3\), bobot \(w_1, w_2, w_3\), bias \(b\), dan fungsi aktivasi sigmoid \(\sigma(z) = \frac{1}{1 + e^{-z}}\). Maka perhitungan output neuron sederhana adalah:

```python
# Contoh Perhitungan dengan Python
import math

# Input dan bobot
x1, x2, x3 = 0.5, 0.3, 0.2
w1, w2, w3 = 0.8, 0.4, 0.1
b = 0.2

# Menghitung z
z = (w1 * x1) + (w2 * x2) + (w3 * x3) + b

# Fungsi aktivasi sigmoid
def sigmoid(z):
    return 1 / (1 + math.exp(-z))

# Output neuron
output = sigmoid(z)
print(f"Nilai z: {z}")
print(f"Output: {output}")
```

Penjelasan:
1. Hitung nilai \(z\):
   \[
z = w_1x_1 + w_2x_2 + w_3x_3 + b
\]

2. Terapkan fungsi aktivasi:
   \[
output = \sigma(z) = \frac{1}{1 + e^{-z}}
\]

Contoh Perhitungan Manual:
- \(x_1 = 0.5, x_2 = 0.3, x_3 = 0.2\)
- \(w_1 = 0.8, w_2 = 0.4, w_3 = 0.1\)
- \(b = 0.2\)

Hitung:
- \(z = (0.8 \times 0.5) + (0.4 \times 0.3) + (0.1 \times 0.2) + 0.2 = 0.66\)
- \(output = \frac{1}{1 + e^{-0.66}} \approx 0.659\)
